## Horizontal polarization

### 14 mm (1.9kW @ 500mA) (7.40 eV)

```python
calculation_folder = Path('SPECTRA_calculations/ReferenceCalculations/')
spectrum = loadSPECTRACalc(calculation_folder/"HPOL_14mm_narrowBDA.dc0")

fundamental = 7.4
FWHM = fundamental/29

plt.figure(figsize=[12,6])

plt.subplot(121)
plt.plot(spectrum['mono_energy'],spectrum['flux'])
plt.plot([fundamental,fundamental],[0,max(spectrum['flux'])],'--',color='black')
plt.xlim([fundamental-5*FWHM,fundamental+5*FWHM])
plt.xlabel("Photon energy")
plt.ylabel("Flux (Photons/s/0.1%BW)")

plt.subplot(122)
plt.plot(spectrum['mono_energy'],spectrum['flux'])
for order in [1,2,3,4]: plt.plot([fundamental*order,fundamental*order],[0,max(spectrum['flux'])],'--',color='black')
plt.xlim([fundamental-5*FWHM,(fundamental*4)+5*FWHM])
plt.xlabel("Photon energy")
plt.ylabel("photons/s/0.1%BW")
plt.show()

plt.figure(figsize=[12,6])
plt.plot(spectrum['mono_energy'],spectrum['flux'])
plt.xlabel("Photon energy")
plt.ylabel("photons/s/0.1%BW")
plt.show()
```

# Wiggler mode

Outside the first harmonic, we operate in 'wiggler' mode (hv > 200 eV, wide open BDA). Which EPU setting is best?

## Calculation details

| Parameter | Value |
|-|-|
|Program | SPECTRA 10.1|
|Mode | Near field>energy dependence>partial flux|
|$\Delta$x,y (mm)| 15 |
|x,y (mm)| 0 |
|Distance from source (m)| 14 |
|Parameter file |84.prm\*|

\*500mA beam current

```python
calculation_folder = Path('SPECTRA_calculations/ReferenceCalculations/')
HPOL_14mm = loadSPECTRACalc(calculation_folder/"HPOL_14mm_openBDA.dc0")
HPOL_15mm = loadSPECTRACalc(calculation_folder/"HPOL_15mm_openBDA.dc0")
HPOL_16mm = loadSPECTRACalc(calculation_folder/"HPOL_16mm_openBDA.dc0")
HPOL_20mm = loadSPECTRACalc(calculation_folder/"HPOL_20mm_openBDA.dc0")
VPOL_14mm = loadSPECTRACalc(calculation_folder/"VPOL_14mm_openBDA.dc0")
VPOL_15mm = loadSPECTRACalc(calculation_folder/"VPOL_15mm_openBDA.dc0")
VPOL_20mm = loadSPECTRACalc(calculation_folder/"VPOL_20mm_openBDA.dc0")

plt.figure(figsize=[12,6])
plt.plot(HPOL_14mm['mono_energy'],HPOL_14mm['flux'],label='HPOL 14mm')
plt.plot(HPOL_15mm['mono_energy'],HPOL_15mm['flux'],label='HPOL 15mm')
plt.plot(HPOL_16mm['mono_energy'],HPOL_16mm['flux'],label='HPOL 16mm')
plt.plot(HPOL_20mm['mono_energy'],HPOL_20mm['flux'],label='HPOL 20mm')
plt.plot(VPOL_14mm['mono_energy'],VPOL_14mm['flux'],label='VPOL 14mm')
plt.plot(VPOL_15mm['mono_energy'],VPOL_15mm['flux'],label='VPOL 15mm')
plt.plot(VPOL_20mm['mono_energy'],VPOL_20mm['flux'],label='VPOL 20mm')
plt.xlabel("Photon energy")
plt.ylabel("photons/s/0.1%BW")
#plt.xlim([500,700])
#plt.ylim([0.8e15,1e15])
plt.legend()
plt.show()
```

Consistent with what we see experimentally, when pushing to lower gaps there are diminishing returns below 19mm HPOL. The increase in flux is probably not worth the costs of a higher heatload.

Above 300eV the flux difference for different EPU settings is not more than 10%, so hardly worth bothering. Below that, you could gain 30% by fine tuning the gap.

It seems that VPOL 14mm is actually better than HPOL 19mm for the same heatload

```python
#*****************************************************
def buildEnergySlice(calculationMatrix,fluxCurve,minEnergy,maxEnergy,energyStep,sliceEnergy):
#*****************************************************
    calculationIndex=int((sliceEnergy - minEnergy)/energyStep)
    
    fig=plt.figure(figsize=(10,8))
    plt.subplot(2,1,1)
    calculationImage=calculationMatrix[:,:,calculationIndex]
    plt.imshow(calculationImage,origin='lower',aspect='1',cmap='gray_r',extent=[-8,8,8,-8])
    aperture = patches.RegularPolygon((0,0), numVertices=4, orientation = math.radians(45), color='tab:orange',radius = sqrt(2)*(2.1/2),linestyle='--',fill=False)
    plt.gca().add_patch(aperture)
    plt.xlabel('Horizontal (mm)')
    plt.ylabel('Vertical (mm)')
    
    plt.subplot(2,1,2)
    plt.plot(fluxCurve['mono_energy'],fluxCurve['flux'])
    plt.plot([sliceEnergy,sliceEnergy],[min(fluxCurve['flux']),max(fluxCurve['flux'])])
    fundamental = 11.41
    plt.plot([fundamental,fundamental],[min(fluxCurve['flux']),max(fluxCurve['flux'])],'--',color='tab:green')
    plt.plot([fundamental*2,fundamental*2],[min(fluxCurve['flux']),max(fluxCurve['flux'])],'--',color='tab:green')
    plt.plot([fundamental*2.72,fundamental*2.72],[min(fluxCurve['flux']),max(fluxCurve['flux'])],'--',color='tab:green')
    plt.plot([fundamental*3,fundamental*3],[min(fluxCurve['flux']),max(fluxCurve['flux'])],'--',color='tab:green')
    plt.xlabel('Energy (eV)')
    plt.ylabel('Flux')
    plt.xlim([minEnergy,maxEnergy])
    
    plt.tight_layout()
    
#*****************************************************
def ShowImage(calculationMatrix,fluxCurve,minEnergy,maxEnergy,energyStep):
#*****************************************************   
    sliceNumberSlider=widgets.FloatSlider(
        value=minEnergy,
        min=minEnergy,
        max=maxEnergy,
        step=energyStep,
        description='Energy:',
        continuous_update=False,
        layout=Layout(width='800px')
    )    
           
    w = interactive(
        buildEnergySlice,
        calculationMatrix=fixed(calculationMatrix),
        fluxCurve=fixed(fluxCurve),
        minEnergy=fixed(minEnergy),
        maxEnergy=fixed(maxEnergy),
        energyStep=fixed(energyStep),
        sliceEnergy=sliceNumberSlider
        
    )
   
    # prevent flicker as the cell size goes to zero when redrawing
    output = w.children[-1]
    output.layout.height = '700px'
    return w 
```

It is possible to shift the calculated spectra around by rescaling the magnetic field data. If you want to go down that road, here is an example code block that will do it for you:

```python
scaleFactor = 0.97

x=[]
y=[]
z=[]

with open("/home/craig/SPECTRA/BLOCH_spectra/Bxy_VP/Gap20_VPp_BxBy_Scan13C.txt",'r') as f:
    for line in f:
        a,b,c = line.rstrip('\n').split('\t')
        x.append(float(a))
        y.append(float(b))
        z.append(float(c))
      
with open("/home/craig/SPECTRA/BLOCH_spectra/Bxy_VP/Gap20_VPp_BxBy_Scan13C__corrected.txt",'w') as f:
    for a,b,c in zip(x,y,z):
        f.write("{:.3f}\t{:.5e}\t{:.5e}\n".format(a,b*scaleFactor,c*scaleFactor))
```
